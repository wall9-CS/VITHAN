```python
import random

random.seed(0)

# 전체 데이터 읽기
with open('/content/drive/MyDrive/code/VTHNKG-OA_NT/discrete_triplets.txt', 'r', encoding='utf-8') as f:
    discrete_triplets = f.readlines()

with open('/content/drive/MyDrive/code/VTHNKG-OA_NT/numeric_triplets.txt', 'r', encoding='utf-8') as f:
    numeric_triplets = f.readlines()

# 전체 데이터 섞기
random.shuffle(discrete_triplets)
random.shuffle(numeric_triplets)
# 각 데이터셋을 파일로 저장
def save_dataset(filename, data):
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(data)

def split(triplets):
    # 데이터셋 분할 비율 설정 (예: train 70%, valid 15%, test 15%)
    total_lines = len(triplets)
    train_split = int(total_lines * 0.8)
    valid_split = train_split + int(total_lines * 0.1)

    # 데이터셋 분할
    train_data = triplets[:train_split]
    valid_data = triplets[train_split:valid_split]
    test_data = triplets[valid_split:]

    return train_data, valid_data, test_data

train_data, valid_data, test_data = split(discrete_triplets)
numeric_split = split(numeric_triplets)
train_data += numeric_split[0]
valid_data += numeric_split[1]
test_data += numeric_split[2]

total_lines = len(train_data) + len(valid_data) + len(test_data)

save_dataset('/content/drive/MyDrive/code/VTHNKG-OA_NT/train.txt', train_data)
save_dataset('/content/drive/MyDrive/code/VTHNKG-OA_NT/valid.txt', valid_data)
save_dataset('/content/drive/MyDrive/code/VTHNKG-OA_NT/test.txt', test_data)

# 데이터셋 크기 출력
print(f"Total triplets: {total_lines}")
print(f"Train set: {len(train_data)} triplets ({len(train_data)/total_lines*100:.2f}%)")
print(f"Valid set: {len(valid_data)} triplets ({len(valid_data)/total_lines*100:.2f}%)")
print(f"Test set: {len(test_data)} triplets ({len(test_data)/total_lines*100:.2f}%)")
```
